```python
#This function calculates GC_content from a DNA string
def GC_content(DNA):
    DNA1 = DNA.translate( { ord(i): None for i in 'TA'} )
    return round(len(DNA1)/len(DNA)*100,6)


def maximum_GC_content(text):
    #This part transforms FASTA format to a list
    text = text.replace('\n','').split(sep='>')
    text.pop(0)

    #This part makes a dictionary for each ID with its GC-content
    dictionary = {}
    for i in text:
        dictionary[i[0:13]] = GC_content(i[13:])
    
    value = max(dictionary.values())
    key = [i for i in dictionary if dictionary[i] == value]
    return print(key[0], value)



maximum_GC_content(example)
```
